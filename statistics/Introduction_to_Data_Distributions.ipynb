# Introduction to Data Distributions

## Exploring Data Distributions with the Titanic Dataset

Data distribution is a crucial concept in data analysis, as it describes how values in a dataset are spread or scattered across a range. By studying data distributions, we can uncover meaningful patterns, measure variability, and gain insights into the characteristics of a dataset. In this example, we utilize the **Titanic dataset**, a well-known collection of passenger information, to explore various types of data distributions.

The Titanic dataset provides both **numerical data** (e.g., age, fare) and **categorical data** (e.g., survival status, passenger class). It offers a rich opportunity to analyze and visualize distributions through histograms, KDE (Kernel Density Estimation) plots, and box plots. By examining these distributions, we can understand key features of the data, such as:
- Central tendencies (mean, median, mode),
- Spread or dispersion (range, variance, standard deviation),
- Patterns or trends within different groups (e.g., survivors vs. non-survivors).

Using Python, we delve into the Titanic dataset to analyze the distribution of attributes like **age**, **fare**, and **class**, demonstrating how data distribution supports informed decisions and predictive modeling. This case study illustrates not only the power of data visualization but also how understanding distributions provides a foundation for advanced statistical analysis. Letâ€™s dive into these examples!

### Load and Preprocess the Dataset

First, we'll import the Titanic dataset using seaborn and handle any missing values to prepare it for analysis.


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Preview the dataset
print(titanic.head())

# Drop rows with missing values for simplicity
titanic_cleaned = titanic.dropna()

print("\nCleaned Data Summary:")
print(titanic_cleaned.info())
```

### Distribution of Age (Numerical Variable - Continuous Data)
Analyze the distribution of passengers' ages. We'll use a histogram with a KDE (Kernel Density Estimate) to observe its pattern.


```python
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_cleaned['age'], kde=True, bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Observation:
* We can identify the central tendency of age, check for skewness, and observe the distribution's shape (e.g., normal, skewed).


### Survival Distribution (Categorical Variable - Discrete Data)
Explore the binary distribution of the survival status (0 = did not survive, 1 = survived).


```python
# Survival Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='survived', data=titanic_cleaned, palette='Set2')
plt.title('Survival Distribution')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

Observation:
* This categorical distribution highlights the proportion of survivors and non-survivors.


### Fare Distribution (Numerical Variable - Continuous Data)
Analyze the fare paid by passengers and observe its distribution using a log scale, as fares often have long tails.


```python
# Distribution of Fare
plt.figure(figsize=(10, 6))
sns.histplot(titanic_cleaned['fare'], kde=True, bins=30, color='green', edgecolor='black')
plt.title('Distribution of Passenger Fare')
plt.xlabel('Fare (in $)')
plt.ylabel('Frequency')
plt.show()
```

Observation:
* Fares may exhibit a skewed distribution due to outliers (e.g., wealthy passengers in 1st class paying higher fares).


### Age Distribution by Survival (Numerical Data Split by Categorical)
Compare age distributions for passengers who survived versus those who didn't.


```python
# Age Distribution by Survival
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic_cleaned, x='age', hue='survived', fill=True, common_norm=False, palette='coolwarm', alpha=0.5)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

Observation:
* This plot shows how age relates to survival, highlighting differences between the two groups.


### Class Distribution (Categorical Variable - Multiclass)
Analyze the distribution of passengers across different ticket classes.


```python
# Class Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='class', data=titanic_cleaned, palette='Set3', order=['First', 'Second', 'Third'])
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

Observation:
* The count of passengers in 1st, 2nd, and 3rd classes helps us understand socioeconomic diversity.


## Exploring Key Characteristics of Data Distributions with the Titanic Dataset

Understanding the **key characteristics of data distributions** is fundamental to analyzing and interpreting data effectively. Data distributions help us uncover patterns, identify central tendencies, measure variability, and visualize trends across datasets. They form the foundation for statistical analysis, predictive modeling, and decision-making.

In this example, we use the **Titanic dataset**, a historical and widely-used dataset in data science, to explore these characteristics. The Titanic dataset contains information about passengers, including their survival status, age, fare, gender, and class. By analyzing both **numerical variables** (e.g., age, fare) and **categorical variables** (e.g., survival status, passenger class), we can examine the datasetâ€™s key features such as:
- **Central Tendency**: Measures of the "center" of the data (mean, median, mode).
- **Dispersion**: The spread or variability of data points (range, standard deviation, variance).
- **Shape**: The pattern and symmetry of data distribution (e.g., normal, skewed, or uniform).
- **Outliers**: Extreme values that deviate significantly from the rest of the data.
- **Frequency**: How often values or ranges of values occur, especially in categorical variables.

Using Python, we visualize these characteristics with histograms, KDE plots, box plots, and count plots to provide actionable insights. The Titanic dataset serves as a compelling example for studying data distributions, combining both continuous and discrete data types to demonstrate the versatility of distribution analysis. Letâ€™s dive deeper into the exploration!

**Central Tendency**
* Mean, Median, and Mode provide insights into the "center" of numerical variables such as age and fare.


```python
# Calculate central tendency for age
mean_age = titanic_cleaned['age'].mean()
median_age = titanic_cleaned['age'].median()
mode_age = titanic_cleaned['age'].mode()[0]
print(f"Mean Age: {mean_age:.2f}")
print(f"Median Age: {median_age:.2f}")
print(f"Mode Age: {mode_age}")
```

**Dispersion**
* Range, Variance, and Standard Deviation measure how spread out the values are.

```python
# Calculate dispersion for fare
min_fare = titanic_cleaned['fare'].min()
max_fare = titanic_cleaned['fare'].max()
range_fare = max_fare - min_fare
std_dev_fare = titanic_cleaned['fare'].std()
print(f"Minimum Fare: {min_fare:.2f}")
print(f"Maximum Fare: {max_fare:.2f}")
print(f"Range of Fare: {range_fare:.2f}")
print(f"Standard Deviation of Fare: {std_dev_fare:.2f}")
```

**Shape of Distributions**
* Assess the overall pattern and symmetry (e.g., normal, skewed, uniform).
* Use histograms and KDE plots for visualization.


```python
# Visualize fare distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic_cleaned['fare'], kde=True, bins=30, color='green', edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare ($)")
plt.ylabel("Frequency")
plt.show()
```

**Observation:** Fares often exhibit a right-skewed distribution due to a few high-paying passengers.

**Frequency**

For categorical variables, we can observe value frequencies, such as survival status or passenger class.

```python
# Frequency of passenger classes
class_counts = titanic_cleaned['class'].value_counts()
print("Passenger Class Frequency:")
print(class_counts)
```

```python
# Visualize class distribution
sns.countplot(x='class', data=titanic_cleaned, palette='Set2', order=['First', 'Second', 'Third'])
plt.title("Passenger Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

**Outliers**

Identify and visualize outliers, which are extreme values that deviate from the rest of the data.

```python
# Detect outliers in fare using box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x=titanic_cleaned['fare'], color='orange')
plt.title("Outliers in Fare")
plt.xlabel("Fare")
plt.show()
```

**Observation:** Fares in the higher class (1st Class) may have extreme values compared to lower classes.


## Introduction to scipy.stats package

The `scipy.stats` subpackage is a powerful and versatile tool for statistical analysis in Python, offering over 80 probability distributions, statistical tests, descriptive metrics, and tools for data fitting. Its key strengths lie in its efficiency, flexibility, and precision, making it suitable for domains like data science, finance, and research.

With support for parametric and non-parametric tests, random variate generation, hypothesis testing, and seamless integration with NumPy and Pandas, `scipy.stats` simplifies complex statistical computations. Its high performance ensures reliability even for large datasets, and its intuitive design makes it accessible for both beginners and experts.

Example applications include:
- **Hypothesis Testing:** Comparing means or testing independence of variables.
- **Probability Modeling:** Analyzing uncertainty with distributions like binomial, Poisson, and Pareto.
- **Descriptive Statistics:** Calculating skewness, kurtosis, and other metrics.
- **Simulations:** Generating random data for experiments or modeling.

In short, `scipy.stats` is an essential package for accurate, efficient, and user-friendly statistical analysis.

### 1. **Bernoulli Distribution (`bernoulli`)**
- **Syntax:** `bernoulli.pmf(k, p)` / `bernoulli.rvs(p, size)`
  - **`k`**: Integer, the outcome (0 or 1).
  - **`p`**: Probability of success float, $(0 \leq p \leq 1)$.
  - **`size`**: Number of random samples to generate (optional).

```python
from scipy.stats import bernoulli

p = 0.6  # Probability of success

# Generate random samples
samples = bernoulli.rvs(p, size=10)
# PMF for success (k=1) and failure (k=0)
pmf_success = bernoulli.pmf(1, p)
pmf_failure = bernoulli.pmf(0, p)
# CDF for success (k=1)
cdf = bernoulli.cdf(1, p)

print("Samples:", samples)
print("PMF Success:", pmf_success)
print("PMF Failure:", pmf_failure)
print("CDF:", cdf)
```

### 2. **Binomial Distribution (`binom`)**
- **Syntax:** `binom.pmf(k, n, p)` / `binom.rvs(n, p, size)`
  - **`k`**: Integer, the number of successes.
  - **`n`**: Number of trials (positive integer).
  - **`p`**: Probability of success for each trial $(0 \leq p \leq 1)$.
  - **`size`**: Number of random samples to generate (optional).


```python
from scipy.stats import binom

n, p = 10, 0.5  # Number of trials and probability of success

# Generate random samples
samples = binom.rvs(n, p, size=10)
# PMF for 4 successes
pmf = binom.pmf(4, n, p)
# CDF for 4 or fewer successes
cdf = binom.cdf(4, n, p)
# PPF for the 95th percentile
ppf = binom.ppf(0.95, n, p)

print("Samples:", samples)
print("PMF:", pmf)
print("CDF:", cdf)
print("PPF (95th Percentile):", ppf)
```

### 3. **Poisson Distribution (`poisson`)**
- **Syntax:** `poisson.pmf(k, mu)` / `poisson.rvs(mu, size)`
  - **`k`**: Integer, the number of events.
  - **`mu`**: Mean event rate (float, positive).
  - **`size`**: Number of random samples to generate (optional).

```python
from scipy.stats import poisson

mu = 3  # Mean event rate

# Generate random samples
samples = poisson.rvs(mu, size=10)
# PMF for exactly 2 events
pmf = poisson.pmf(2, mu)
# CDF for 2 or fewer events
cdf = poisson.cdf(2, mu)
# PPF for the 90th percentile
ppf = poisson.ppf(0.9, mu)

print("Samples:", samples)
print("PMF:", pmf)
print("CDF:", cdf)
print("PPF (90th Percentile):", ppf)
```

### 4. **Multinomial Distribution (`multinomial`)**
- **Syntax:** `multinomial.pmf(x, n, pvals)` / `multinomial.rvs(n, pvals)`
  - **`x`**: Array-like, the number of occurrences for each outcome.
  - **`n`**: Number of trials (positive integer).
  - **`pvals`**: List of probabilities for each outcome, must sum to 1.
  - **`size`**: Number of random samples to generate (optional).

```python
from scipy.stats import multinomial

n = 10
pvals = [0.6, 0.3, 0.1]  # Probabilities of outcomes

# Generate random samples
samples = multinomial.rvs(n, pvals, size=5)
# PMF for a specific outcome set
pmf = multinomial.pmf([5, 3, 2], n, pvals)

print("Samples:", samples)
print("PMF:", pmf)
```

### 5. **Uniform Distribution (`uniform`)**
- **Syntax:** `uniform.pdf(x, loc, scale)` / `uniform.rvs(loc, scale, size)`
  - **`x`**: Float or array-like, the value(s) for which to evaluate the function.
  - **`loc`**: Lower bound of the distribution (default is 0).
  - **`scale`**: Width of the distribution (default is 1).
  - **`size`**: Number of random samples to generate (optional).

```python
from scipy.stats import uniform

loc, scale = 0, 1  # Range of the distribution

# Generate random samples
samples = uniform.rvs(loc=loc, scale=scale, size=10)
# PDF at x = 0.5
pdf = uniform.pdf(0.5, loc=loc, scale=scale)
# CDF at x = 0.5
cdf = uniform.cdf(0.5, loc=loc, scale=scale)
# PPF for the 75th percentile
ppf = uniform.ppf(0.75, loc=loc, scale=scale)

print("Samples:", samples)
print("PDF:", pdf)
print("CDF:", cdf)
print("PPF (75th Percentile):", ppf)
```

### 6. **Normal Distribution (`norm`)**
- **Syntax:** `norm.pdf(x, loc, scale)` / `norm.rvs(loc, scale, size)`
  - **`x`**: Float or array-like, the value(s) for which to evaluate the function.
  - **`loc`**: Mean of the distribution (default is 0).
  - **`scale`**: Standard deviation of the distribution (default is 1).
  - **`size`**: Number of random samples to generate (optional).


```python
from scipy.stats import norm

mean, std_dev = 0, 1  # Mean and standard deviation

# Generate random samples
samples = norm.rvs(loc=mean, scale=std_dev, size=10)
# PDF at x = 0
pdf = norm.pdf(0, loc=mean, scale=std_dev)
# CDF at x = 0
cdf = norm.cdf(0, loc=mean, scale=std_dev)
# PPF for the 97.5th percentile
ppf = norm.ppf(0.975, loc=mean, scale=std_dev)

print("Samples:", samples)
print("PDF:", pdf)
print("CDF:", cdf)
print("PPF (97.5th Percentile):", ppf)
```

### 7. **Exponential Distribution (`expon`)**
- **Syntax:** `expon.pdf(x, scale)` / `expon.rvs(scale, size)`
  - **`x`**: Float or array-like, the value(s) for which to evaluate the function.
  - **`scale`**: Mean (inverse rate parameter \(\lambda = 1/\text{scale}\), default is 1).
  - **`size`**: Number of random samples to generate (optional).


```python
from scipy.stats import expon

scale = 2  # Mean of the distribution

# Generate random samples
samples = expon.rvs(scale=scale, size=10)
# PDF at x = 1
pdf = expon.pdf(1, scale=scale)
# CDF at x = 1
cdf = expon.cdf(1, scale=scale)
# PPF for the 80th percentile
ppf = expon.ppf(0.8, scale=scale)

print("Samples:", samples)
print("PDF:", pdf)
print("CDF:", cdf)
print("PPF (80th Percentile):", ppf)
```

### 8. **Log-Normal Distribution (`lognorm`)**
- **Syntax:** `lognorm.pdf(x, s, loc, scale)` / `lognorm.rvs(s, loc, scale, size)`
  - **`x`**: Float or array-like, the value(s) for which to evaluate the function.
  - **`s`**: Shape parameter (standard deviation of the log of the variable).
  - **`loc`**: Mean of the log-transformed variable (default is 0).
  - **`scale`**: Scale parameter (\(e^{\text{mean}}\), default is 1).
  - **`size`**: Number of random samples to generate (optional).

```python
from scipy.stats import lognorm

shape, scale = 0.5, 1  # Shape and scale parameters

# Generate random samples
samples = lognorm.rvs(shape, scale=scale, size=10)
# PDF at x = 1
pdf = lognorm.pdf(1, shape, scale=scale)
# CDF at x = 1
cdf = lognorm.cdf(1, shape, scale=scale)
# PPF for the 90th percentile
ppf = lognorm.ppf(0.9, shape, scale=scale)

print("Samples:", samples)
print("PDF:", pdf)
print("CDF:", cdf)
print("PPF (90th Percentile):", ppf)
```

### 9. **Pareto Distribution (`pareto`)**
- **Syntax:** `pareto.pdf(x, b, loc, scale)` / `pareto.rvs(b, loc, scale, size)`
  - **`x`**: Float or array-like, the value(s) for which to evaluate the function.
  - **`b`**: Shape parameter (tail index, positive).
  - **`loc`**: Location parameter (default is 0).
  - **`scale`**: Scale parameter (default is 1).
  - **`size`**: Number of random samples to generate (optional).

```python
from scipy.stats import pareto

b = 2.5  # Shape parameter

# Generate random samples
samples = pareto.rvs(b, size=10)
# PDF at x = 1
pdf = pareto.pdf(1, b)
# CDF at x = 1
cdf = pareto.cdf(1, b)
# PPF for the 95th percentile
ppf = pareto.ppf(0.95, b)

print("Samples:", samples)
print("PDF:", pdf)
print("CDF:", cdf)
print("PPF (95th Percentile):", ppf)
```
