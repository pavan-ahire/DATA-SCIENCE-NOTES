# Sampling_technique


```python
import pandas as pd
import numpy as np
import seaborn as sns

tips=sns.load_dataset('tips')

data={
    'total_bill':tips.total_bill,
    'tip':tips.tip,
    'sex':tips.sex
}
df=pd.DataFrame(data)

print("original dataset \n",df.head())
```

```python
n_samples=5
random_indices_np =np.random.choice(df.index,size=n_samples,replace=False)
simple_random_sample_np=df.loc[random_indices_np]
print("\n simple random sampling using np.choose: \n",simple_random_sample_np)
```

```python
simple_random_sample_df=df.sample(n=n_samples,random_state=42)
print("\n simple random sampling using dataframe.sample: \n",simple_random_sample_df)
```

## systematic sampling

```python
n_step=50
shuffled_df=df.sample(frac=1,random_state=42).reset_index(drop=True)
systematic_sample=shuffled_df.iloc[::n_step]
print("\n systematic sampling using shuffle and slicing: \n ", systematic_sample)
```

```python
## stratified sampling

strate='sex'
stratified_sample= df.groupby(strate, group_keys=False).apply(lambda x:x.sample(n=5,random_state=42))
print("\n stratified sampling using groupby : \n",stratified_sample)
```
