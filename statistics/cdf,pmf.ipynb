## Bernoulli Distribuition

```python
from scipy.stats import bernoulli

p = 0.6

samples = bernoulli.rvs(p, size=10)
print(samples)

pmf_success = bernoulli.pmf(1, p)
pmf_failure = bernoulli.pmf(0, p)
print(pmf_success , pmf_failure)

cdf_ = bernoulli.cdf(-1,p)
print(cdf_)

# P(X=x) = px(1-p)1-x
# P(X=0) = 0.6 x (0.4)1-0   #(anything raise to 0 = 1)
```

## Binomial Distribuition

```python
# 0.5 ex
```

```python
from scipy.stats import binom

p=0.5
n=10
samples = binom.rvs(n,p, size=10)
print(samples)

pmf_binom = binom.pmf(4,n,p)
print(pmf_binom)

cdf_ = binom.cdf(4,n,p)
print(cdf_)
```

## Poision Distribuition

```python
from scipy.stats import poisson

# P(X=x)=lambda x....

mu = 3
samples = poisson.rvs(mu, size=10)
print(samples)

pmf = poisson.pmf(2,mu)
print(pmf)

cdf = poisson.cdf(2,mu)
print(cdf_)

ppf = poisson.ppf(0.9,mu)
print(ppf)
```

## Multinomial Distribuition

```python
from scipy.stats import multinomial
n = 10
pvals = [0.6 ,0.3 , 0.1]

samples = multinomial.rvs(n, pvals, size=5)
print(samples)

pmf = multinomial.pmf([3,6,1] ,n , pvals)
print(pmf)
```

## Uniform Distribuition

```python
from scipy.stats import uniform
loc=0
scale=1
samples = uniform.rvs(loc = loc,scale=scale,size=10)
samples

pdf = uniform.pdf(0.5,loc=0,scale=scale)
print(pdf)

cdf = uniform.cdf(0.5,loc=0,scale=scale)
print(cdf)

ppf = uniform.ppf(0.75, loc=loc , scale=scale)
print(ppf)
```

## Normal Distribuition (norm)

```python
# bell curve
```

```python
from scipy.stats import norm

mean = 0
std = 1

samples = norm.rvs(loc = mean , scale = std , size = 10)
print(samples)

pdf = norm.pdf(0, loc = mean , scale = std)
print(pdf)

cdf_ = norm.cdf(0, loc = mean , scale=std)
print(cdf_)

ppf_ = norm.ppf(0.5, loc=mean,scale=std)
print(ppf_)
```

# Exponential Distribuition (expon)

```python
from scipy.stats import expon

scale = 2
samples = expon.rvs(scale = scale , size = 10)
print(samples)

pdf = expon.pdf(1 , scale = scale)
print(pdf)

cdf = expon.cdf(1,scale=scale)
print(cdf)

ppf = expon.cdf(0.8, scale=scale)
print(ppf)
```

```python
from scipy.stats import lognorm

shape = 0.5
scale = 1

samples = lognorm.rvs(shape, scale=scale , size=10)
print(samples)

pdf = lognorm.pdf(1, shape, scale= scale)
print(pdf)

cdf_ = lognorm.cdf(1, shape, scale=scale)
print(cdf_)

ppf = lognorm.ppf(0.9, shape, scale=scale)
print(ppf)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, probplot

tips = sns.load_dataset('tips')
print(tips.head())

print(tips['total_bill'].mean(),tips['total_bill'].std())

total_bill_th = norm.rvs(tips['total_bill'].mean(), scale= tips['total_bill'].std(), size=len(tips))
print(total_bill_th)

fis, ax = plt.subplots(1,3, figsize = (12,4))
plt.subtitle("Therotical total nill data")
sns.histplot(tips['total_bill_th'], ax = ax[0])
sns.kdeplot(tips['total_bill_th'], ax = ax[1])
sns.boxplot(tips['total_bill_th'], ax = ax[2])
plt.tight_layout()
plt.show()

fis, ax = plt.subplots(1,3, figsize = (12,4))
plt.subtitle("actual total bill data")
sns.histplot(tips['total_bill_th'], ax = ax[0])
sns.kdeplot(tips['total_bill_th'], ax = ax[1])
sns.boxplot(tips['total_bill_th'], ax = ax[2])
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

fig , ax = plt.subplots(1,2)
print(ax)
stats.probplot(tips['total_bill'],dist = stats.norm,plot=ax[0])
stats.probplot(total_bill_th, dist = stats.norm, plot=ax[1])

ax[0].set_title["total bill QQ plot"]
ax[1].set_title["total bill theroretical QQ plot"]
plt.show()
```
