```python
import scipy.stats
```

```python

# Probability of success (Head = 1)
p = 0.7

# PMF calculations
pmf_0 = 1 - p
pmf_1 = p         # P(X = 1)

# CDF calculations
# CDF(0) = P(X ≤ 0) = P(X = 0)
cdf_0 = pmf_0

# CDF(1) = P(X ≤ 1) = P(X = 0) + P(X = 1)
cdf_1 = pmf_0 + pmf_1

print("PMF:")
print("P(X = 0) =", pmf_0)
print("P(X = 1) =", pmf_1)

print("\nCDF:")
print("CDF(0) =", cdf_0)
print("CDF(1) =", cdf_1)
```

# Bernoulli Distribuition

```python
from scipy.stats import bernoulli

p = 0.6

samples = bernoulli.rvs(p, size=10)
print(samples)

pmf_success = bernoulli.pmf(1, p)
pmf_failure = bernoulli.pmf(0, p)
print(pmf_success , pmf_failure)

cdf_ = bernoulli.cdf(-1,p)
print(cdf_)

# P(X=x) = px(1-p)1-x
# P(X=0) = 0.6 x (0.4)1-0   #(anything raise to 0 = 1)
```
