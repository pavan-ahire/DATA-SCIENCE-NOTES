**6. Create a class BankAccount with attributes account_holder balance. Write method to diposite and withdraw money from the account. Create the object and perform the deposite and withdrawal action.**

```python
class BankAccount:
    def __init__(this,account_holder,balance):
        this.account_holder = account_holder
        this.balance = balance

    def deposite_amount(this,amount):
        this.amount = amount
        print(f"Total Balance is: {this.balance}")
        this.balance +=this.amount
        print(f"The deposited amount is {this.amount} and total balance is:",this.balance)

    def withdraw_amount(this,amount):
        this.amount = amount
        print(f"The withdrawal amount is {this.amount} and available balance is:",this.balance - this.amount)

obj = BankAccount("pavan",50000)
obj.deposite_amount(10000)
obj.withdraw_amount(20000)
```

**7. Write a class circle with the attribute radius: Add a method to calculate the circumference of the circle. use the formula: circumference = 2*pi * radius. Create an object an test the mothod.**

```python
import math
class circle:
    def __init__(this,radius):
        this.radius = radius

    def cal_circum_of_circle(this):
        print(f"Radius is {this.radius} and circumference of circle is:",2* math.pi*this.radius)

obj = circle(7)
obj.cal_circum_of_circle()
```

**9. Create a class student that accept the student name,age and grade in the constructor. Add a method is_passed() that checks if the grade is above 50. Create an object and check if the student has passed.**

```python
class student:
    def __init__(this,name,age,grade):
        this.name = name
        this.age = age
        this.grade = grade

    def is_passed(this):
        if this.grade > 50:
            print(f"{this.name} is Passed...")
        else:
            print(f"{this.name} is Failed...")

    def display(this):
        print(f"\nName: {this.name} \nAge: {this.age} \nGrade: {this.grade}")

obj  = student("Devendra",23,60)
obj.is_passed()
obj.display()
```
