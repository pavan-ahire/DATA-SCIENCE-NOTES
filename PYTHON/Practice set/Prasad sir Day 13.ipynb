- Q1 (Medium): Write a function that takes a list of strings and returns a new list with all strings converted to
uppercase and sorted alphabetically.

```python
def str1(n):
    a = sorted(n)
    lis = []
    for v in a:
        lis.append(v.upper())
    print(lis)

n=["vaibhav","pavan","devendra","amol"]
str1(n)
```

```python
#2. Control Flow
#Q2 (Medium): Write a Python program that checks whether a number is a prime or not (input from user).

def checker(n):
    if n<=1:
        print("Not a prime number")
        return
    for i in range(2,n):
        if n%i==0:
            print("Not Prime")
            break
    else:
        print("Prime")
n = int(input("Enter the number:"))
checker(n)
```

```python
#6. Tuples
#Q6 (Medium): Write a program that takes a list of tuples (each containing name and score), and returns the
# name of the student with the highest score.

l=[('vaibhav',99),('dev',67),('pavan',90),('milind',69)]
highscore=0
name=""
for sname,score in l:
    if score> highscore:
        highscore=score
        name=sname
print("high score is :",highscore)
print("name is :",name)
```

```python
#7. String
# Q7 (madium): write a function that take a string and return a new string with all vowels removed.

def remove_vowels():
    str1 = "Innomatics Research Lab"
    vowels = "aeiouAEIOU"
    result = ""
    for char in str1:
        if char not in vowels:
            result+=char
    print(result)

remove_vowels()
```
