**Q5 (Medium): Write a Python program to check if a number is a palindrome using a while loop.**

```python
num = int(input("Enter the number:"))
palindrome = num
rev=0
while num>0:
    digit = num%10
    rev=rev*10+digit
    num//=10
if palindrome==rev:
    print("It is palindrome")
else:
    print("Not palindrome")
```

**Q7 (Medium): Write a function that returns the Fibonacci sequence up to a given number of terms.**

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n=int(input("Enter a number to generate fibonacci series:"))
for i in range(n+1):
    print(fibonacci(i),end=",")
```

**10. List Comprehension & Lambda**
- Q15 (Medium): Using list comprehension and lambda, create a list of all numbers from 1 to 30 that are divisible by
both 3 and 5.

```python
lis = [i for i in range(1,31) if i%3==0 and i%5==0] # Using list comprhension
print(lis)
```

```python
# Method 1st using lambda
divides = [x for x in range(1,31) if (lambda i:i%3==0 and i%5==0)(x)]
divides
```

```python
# Method 2nd using lambda
num = lambda x:x%3==0 and x%5==0
divides = [x for x in range(1,31) if num(x) ]
divides
```

```python
divides = lambda x:x   # Method 3rd using lambda
for x in range(1,31):
    if x%3==0 and x%5==0:
        print(divides(x))
```
