**10. Create s function that takes a number and returns True if its a prime number,otherwise False.**

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True
    
print(is_prime(2))
print(is_prime(4))
print(is_prime(13))
```
