**1. Variables and Data Types**
- Q1 (Easy): Write a Python program to swap two numbers using a temporary variable.

```python
a = 4
b = 5
print("Before Swapping")
print("a:",a)
print("b:",b)
temp=a
a=b
b=temp
print("After Swapping")
print("a:",a)
print("b:",b)
```

**Q5 (Medium): Write a program to find the factorial of a number using a while loop.**

```python
num = int(input("Enter the number:"))
i = 1
fact = 1
while i<=num:
    fact*=i
    i+=1
print(f"Factorial of {num} is : {fact}")
```

**Q7 (Medium): Write a function to check whether a number is prime.**

```python
def check_prime(n):
    if n <= 1:
        return "Not Prime"
    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"
    return "It is prime"
check_prime(7)
```

**Q9 (Medium): Write a function that takes a list and returns a new list with only the unique elements.**

```python
def unique(lis1):
    lis2 = []
    for i in lis1:
        if i not in lis2:
            lis2.append(i)
    return lis2
print(unique([1,1,2,2,3,4,5,5,6]))
```

**7. Strings**
- Q10 (Easy): Write a Python program to count the number of vowels in a string.

```python
str1 = input("Enter the string:")
vowels = "aeiouAEIOU"
count = 0
for i in str1:
    if i in vowels:
        count+=1
print("The number of vowels in that string is:",count)
```

**Q11 (Medium): Write a program to check if a given string is a palindrome.**

```python
def checker(str2):
    if str2==str2[::-1]:
        print("It is palindrome")
    else:
        print("Not palindrome")

checker('mam')
checker('vaibhav')
checker('12321')
```

**Q13 (Medium): Write a program to count the frequency of each character in a string using a dictionary.**

```python
def counter():
    string = input("enter the string:")
    frequency = {}
    for i in string:
        if i in frequency:
            frequency[i]+=1
        else:
            frequency[i]=1
    return frequency
counter()
```

**9. File Handling**
- Q14 (Medium): Write a Python program to read a text file and count the number of lines and words.

```python
# open file for output

try:
    outfile = open("examples.txt","w")

    # write data to the file
    outfile.write("vaibhav umbarkar\n")
    outfile.write("Abraaz\n")
    outfile.write("Nikhil\n")
    outfile.write("pavan\n")

finally:
    outfile.close() # close the output file 
```

```python
try:
    infile = open("examples.txt","r")
    line = 0
    words = 0
    for i in infile:
        print(i.strip())
        line+=1
        word=i.split()
        words+=len(word)
    print("\nNumber of lines:",line)
    print("Number of words:",words)
    print(infile.read())

finally:
    infile.close()
```

**10. List Comprehension & Lambda**
- Q15 (Medium): Using list comprehension, create a list of squares of all even numbers from 1 to 20.

```python
# list comprehensions
lis = [i**2 for i in range(1,21) if i%2==0]
lis
```

```python
square=lambda i:i**2
for i in range(1,21):
    if i%2==0:
        print(square(i),end=" , ")
```
