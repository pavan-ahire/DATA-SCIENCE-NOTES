```python
#9. write a code to return the number of unique values in a series.
un = pd.Series([1,2,3,2,3,4,5,3,4,6])
lis = []
for i in un:
    if i not in lis:
        lis.append(i)
print(pd.Series(lis))
```

```python
# Method 2
un = pd.Series([1,2,3,2,3,4,5,3,4,6])
s=set(un)
l=list(s)
for i in range(len(l)):
    print(f'{i} {l[i]}')

```
