**3. Employee salary with bonus.**
- Define a class employee with attributes name, base_salary , and bonus percentage. Write a method to calculate total salary.

```python
class Employee:
    def __init__(this,name,base_salary,bonus):
        this.name = name
        this.base_salary = base_salary
        this.bonus = bonus

    def calculate_total_salary(this):
        salary = this.base_salary+(this.base_salary * this.bonus/100)
        print(f"Name: {this.name} \nSalary: {salary} ")

obj = Employee("pavan",45000,10)
obj.calculate_total_salary()
```

**4. Library book sysytem.**
- Design a class Book with attributes like title, author, and availability. Add methods to borrow and return a book. Track how many times a book has been borrowed.

```python
class Book:
    def __init__(this,title,author):
        this.title = title
        this.author = author
        this.availability = True
        this.borrow_count = 0

    def borrow_book(this):
        if this.availability:
            this.availability = False
            
            this.borrow_count+=1
            print(f"\nYou have borrowed {this.title} ,Borrow {this.borrow_count} times")
        else:
            print(f"\n{this.title} currently not available")

    def return_book(this):
        if not this.availability:
            this.availability =  True
            print(f"\n{this.title} Has been return")
        else:
            print(f"\n{this.title},Was not borrowed")

    def status(this):
        if this.availability:
            print(f"\n{this.title} is available, Borrowed {this.borrow_count} times")
        else:
            print(f"\n{this.title} is not available, Borrowed {this.borrow_count} times")
        

obj = Book("Pride and prejudice","Jane Austen")

obj.status()  
obj.borrow_book()
obj.status()
obj.borrow_book()
obj.status()
obj.return_book()
obj.status()
```

**5 Rectangle class with coparision.**
- Create a rectangle class with width and height. Add a method to calculate area and overload the > operator to compare rectangle by area.

```python
class Rectangle:
    def __init__(this,width,height):
        this.width = width
        this.height = height

    def calculate_area(this):
        return this.width * this.height

    def __gt__(this,other):
        return this.calculate_area() > other.calculate_area()


        

rect1= Rectangle(8,10)
rect2 = Rectangle(3,5)

print("Area of rect1:-",rect1.calculate_area())
print("Area of rect2:-",rect2.calculate_area())

if rect1 > rect2:
    print("Rectangle rect1 is greater than rect2")
else:
    print("Rectangle rect2 is greater than rect1")

        
```

**6. Inheritance example.**
- Create a base class shape and derived class circle and square. Add method in each to calculate the area.

```python
import math

class shape:
    def area(this):
        pass

class circle(shape):
    def __init__(this,radius):
        this.radius = radius

    def area(this):
        return math.pi * this.radius **2

class square(shape):
    def __init__(this,side):
        this.side = side


    def area(this):
        return this.side **2

C = circle(8)
S = square(8)

print("Area of circle is:",C.area())
print("Area of square is:",S.area())
        
```

**7. inventory management system.**
- Create a class product with name,price and quantity. write method to update stock and calculate the total value of the stock.

```python
class product:
    def __init__(this,name,price,quantity):
        this.name = name
        this.price = price
        this.quantity = quantity

    def set_stock(this,amount):
        this.quantity += amount
        print(f"\nUpdated quantity is: {this.quantity}")

    def total_stock_value(this):
        this.price *= this.quantity
        return this.price

    def display(this):
        print(f"\n1.Product Name: {this.name} \n2.Price: {this.price} \n3.Quantity: {this.quantity}")

Ob = product("iphones",50000,10)
print("Initial stock value:",Ob.total_stock_value())
Ob.display()
Ob.set_stock(15)
print("Updated stock value is:",Ob.total_stock_value())
print()
Ob.display()

        
```

**8. Car speed simulation.**
- Create a class car with methods to accelerate ,break and display current speed.Make sure the speed doesent go below 0 or above a max limit.

```python
class Car:
    def __init__(self, max_speed):
        self.speed = 0
        self.max_speed = max_speed

    
    def accelerate(self, increase):
        self.speed += increase
        if self.speed > self.max_speed:
            self.speed = self.max_speed
        print(f"Accelerated. Current speed: {self.speed} km/h")

   
    def brake(self, decrease):
        self.speed -= decrease
        if self.speed < 0:
            self.speed = 0
        print(f"Braked. Current speed: {self.speed} km/h")

   
    def display_speed(self):
        print(f"Current speed: {self.speed} km/h")


my_car = Car(150)

my_car.accelerate(50)
my_car.accelerate(120)
my_car.brake(30)
my_car.brake(200)
my_car.display_speed()
```

***9. Class Method and Static Method Example***
- Create a class MathTools with a static method to check if a number is prime and a class method to count how many
times the method has been used.

```python
class MathTools:
    _count = 0  

    @staticmethod
    def is_prime(n):
        MathTools._count+=1
        for i in range(2,n):
            if n % i ==0:
                print(f"{n} is not a prime number")
                break
        else:
            print(f"\n{n} is a prime number")

       

    @classmethod
    def get_count(cls):
        print(f"\nPrime number method is called: {cls._count} times")


p = MathTools()
p.is_prime(4)
p.is_prime(6)
p.is_prime(8)
p.is_prime(9)
p.is_prime(2)
p.is_prime(3)
p.is_prime(5)
p.is_prime(7)
p.get_count()
```
