**1. Even or odd counter**
- Write a program to count how many even and odd numbers are in a given list.

```python
even = 0
odd = 0
n = list(eval(input("Enter a list:")))
for i in n:
    if i%2==0:
        even+=1
    else:
        odd+=1

print("Even:",even)
print("Odd:",odd)
```

**2. Factorial Using loop**
- Ask the user for a number and calculate its factorial using a for or while loop.

```python
fact = 1
n = int(input("Enter the number:"))
for i in range(1,n+1):
    fact = fact*i
print(fact)
```

**3. Number Range with condition.**
- print all numbers from 1 to 50 , but skip multiple of 5 and stop the loop if the number reaches 35.

```python
for i in range(1,50):
    if i%5==0:
        continue
    elif i>=35:
        break
    else:
        print(i)
```

**4. simple calculator(if-elif).**
- write a program that asks for two numbers and an operator(+,-,*,/) and perform the operation using conditional statements.

```python
a = int(input("Enter the first number:"))
b = int(input("Enter the second number:"))
c = input("enter the operation that you want to perform:")
if c=="+":
    print("Additon is :",a+b)
elif c=="-":
    print("Substraction is :",a-b)
elif c=="*":
    print("Multiplication is:",a*b)
elif c=="/":
    print("Division is:",a/b)
else:
    print("please enter the correct option (+,-,*,/)")
```

**5. Check leap year.**
- Take a year as a input and check if it is a leap year using conditions.

```python
year = int(input("Enter the year:"))
if (year%4==0) or (year%100==0) and (year%400==0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

**9.Count positive, Negative and zero numbers**
- from a given list of numbers,count how many are positive, negative, or zero using a loop and conditionals

```python
n=list(eval(input("enter number")))
positive=0
negative=0
zero=0
for i in n:
    if i>0:
        positive+=1
    elif i<0:
        negative+=1
    else:
        zero+=1
print(n)
print(f"positive are :{positive}")
print(f"negative are :{negative}")
print(f"zero are :{zero}")
```

**10.Check if a Number is a Palindrome (with strings)**
- check if a number entered by the user is a palindrome using string reversal.

```python
n=input("enter a value")
rev=''.join(reversed(n))
if n==rev:
    print(f"{n} is a palindrom ")
else:
    print(f"{n} is not palindrom")
```
