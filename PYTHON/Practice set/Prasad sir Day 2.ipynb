**8 Reverse the list [1,2,3,4,5] without using reverse() or [::-1].**

```python
a = [1,2,3,4,5]
b=len(a)
print(b)
c=[]
for i in range(b-1,-1,-1):
    c.append(a[i])
print('Reverse of the list ',c)
```
