**1. Write a program that takes a number from the user and checks if it is even or odd.**

```python
n = int(input("Enter the number:"))
if n%2==0:
    print(f"Given number {n} is - Even")
else:
    print(f"Given  number {n} is - Odd")
```

**2. Write a python program that check wheather a given number is positive,negative or zero.**

```python
n = int(input("Enter the number:"))
if n >0:
    print(f"Given number {n} is - Positive")
elif n < 0:
    print(f" Given number {n} is - Negative")
else:
    print(f"Given number {n} is - Zero")
```

**5. Write a program that check whether a given yesr is a leap year or not.**

```python
leap_year = int(input("Enter the year:"))
if leap_year%4==0 or leap_year%100==0 and leap_year%400==0:
    print(f"{leap_year} year is a leap year")
else:
    print(f"{leap_year} year is not leap year")
```

**6. write a program that takes a character from the user and check wheather it is vowel or consonant.**

```python
a  = input("Enter the character:")
b = ('a','e','i','o','u')
if a in b:
    print("The given character is a vowel")
else:
    print("The given character is a Consonant")
```

**8. write a program that take a marks from the user and prints the grade as per the following.**

- 90 and above : A
- 80-89: B
- 70-79: C
- Below 60: F

```python
grade = int(input("Enter the marks:"))
if grade >=90:
    print("Grade is:- A")
elif (grade>=80) and (grade<90):
    print("Grade is:- B")
elif (grade>=70) and (grade<80):
    print("Grade is:- C")
else:
    print("Grade is:- F")
```
