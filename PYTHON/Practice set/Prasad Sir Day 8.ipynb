## Question 3

```python
# factorial using recursion
def fact(num):
    if num == 1:
        return num
    else:
        return num*fact(num-1)
        

num=int(input("Enter a number: "))
print(f"Factorial of {num} is :",fact(num))
```

## Question 6

```python
# checking if the string is palindrome
def is_palindrome(s):
    if s == s[::-1]:
        return True
    else:
        return False

s = input("Enter a stirng : ")
is_palindrome(s)
```

## Question 7

```python
# Fibonacci series
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

n=int(input("Enter a number to generate fibonacci series:"))
for i in range(n+1):
    print(fibonacci(i))
```

```python
# function that returns sum of all digits in the number
num = input('Enter a number: ')
def sum(num):
    sum = 0
    for i in num:
        sum += int(i)
    print('Sum of digits in the number is: ',sum)

sum(num)
```

## Question 9

```python
# checking if a number is prime
def is_prime(n):
    for i in range(2,n):
        if n %i==0:
            return False
    else:
        return True

n=int(input("Enter a number to check whether it is prime number: "))
is_prime(n)
```

```python
# reversing the input string
string = input('Enter a string: ')
def rev(string):
    print('Original string is: ',string)
    print('Reversed string is:',string[::-1])

rev(string)
```

```python
# reversing the input string
string = input('Enter a string: ').split(" ")

def rev(string):
    print('Original string is: ',' '.join(string))
    print('Reversed string is:',' '.join(string[::-1]))

rev(string)
```
