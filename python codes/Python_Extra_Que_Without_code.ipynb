## 3. Write a Python Program to Find Factorial of Number Using Recursion.

````python
# recurssion code
def fact(n):
    if n<=1:
        return n
    else:
        return n*fact(n-1)

n=int(input("enter a number"))
print(fact(n))
````

````python
# without recurssion normal code

n=int(input("enter a number :"))
fact=1
for i in range(1,n+1):
    fact*=i

print(fact)
````

````python
# using exception handeling
try:
    n=int(input("enter a number"))
    fact=1
    for i in range(1,n+1):
        fact*=i
except ValueError as e:
    print("error",e)
else:
    print("Factorial of ",n,"is :",fact)
finally:
    print("code is successfully executed")
````

````python
arr1 = [1, 2, 2, 3]  # Monotonic (non-decreasing)
arr2 = [3, 2, 1]     # Monotonic (non-increasing)
arr3 = [1, 3, 2, 4]  # Not monotonic

def monotonic(a):
    j=sorted(a)
    k=sorted(a,reverse=True)
    if a==j:
        print("\nMonotonic (non-decreasing)")
        print("given array: ",a)
        print("sotred of it ASC :",j)
    elif a==k:
        print("\nMonotonic (non-increasing)")
        print("given array: ",a)
        print("sotred of it DEC :",k)
    else:
        print("\nNot Monotonic")
        print(a)
        
    
````

## 30.  Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

````python
class generater:
    def __init__(self,n):
        self.n=n
    def gen_num(self):
        for i in range(self.n+1):
            if i % 7==0:
                yield i

obj=generater(50)
obj.gen_num()

for i in obj.gen_num():
    print(i)
````

## 41. Create a function that replaces all the vowels in a string with a specified character.
```Examples
replace_vowels("the aardvark", "#")  "th# ##rdv#rk"
replace_vowels("minnie mouse", "?")  "m?nn?? m??s?"
replace_vowels("shakespeare", "*")  "shkspr*"

````python
def replace_vowels(txt, char):
    """
    Replaces all vowels in the input string txt with the given character char.
    """
    vowels = "aeiouAEIOU"
    return ''.join(char if c in vowels else c for c in txt)

print(replace_vowels("the aardvark", "#"))   #  "th# ##rdv#rk"
print(replace_vowels("minnie mouse", "?"))   #  "m?nn?? m??s?"
print(replace_vowels("shakespeare", "*"))    #  "sh*k*sp**r*"
````
