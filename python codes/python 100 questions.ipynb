````python
def gen(n):
    ans = []

    def f(s, o, c):
        if o == n and c == n:
            ans.append(s)
            return

        if o < n:
            f(s + "(", o + 1, c)

        if c < o:
            f(s + ")", o, c + 1)

    f("", 0, 0)
    return ans


print(gen(3))
````

## 30.  Define a class with a generator which can iterate the numbers, which are divisible by 7, between a given range 0 and n.

````python
class generator:
    def __init__(s,rs,re):
        s.rs=rs
        s.re=re
    def m(s):
        for i in range(s.rs,s.re):
            if i%7==0:
                yield i

o=generator(0,100)
for i in o.m():
    print(i)
````

## 32. Define a class Person and its two child classes: Male and Female. All classes have a method "getGender" which can print "Male" for Male class and "Female" for Female class.

````python
# using abstract method.

from abc import ABC, abstractmethod

class person(ABC):
    def __init__(s):
        pass

    @abstractmethod
    def getgender(s):
        pass

class male(person):
    def __init__(s):
        pass 
    def getgender(s):
        print('male')

class female(person):
    def __init__(s):
        pass
    def getgender(s):
        print("female")
om=male()
of=female()
om.getgender()
of.getgender()
````

````python
def gen(n):
    ans = []

    def f(s, o, c):
        if o == n and c == n:
            ans.append(s)
            return

        if o < n:
            f(s + "(", o + 1, c)

        if c < o:
            f(s + ")", o, c + 1)

    f("", 0, 0)
    return ans


print(gen(3))
````

## 100. Write a Python program to print all pronic numbers between 1 and 100.
```
A pronic number, also known as an oblong number or rectangular number, is a type of
figurate number that represents a rectangle. It is the product of two consecutive integers, n
and (n + 1). Mathematically, a pronic number can be expressed as:
For example, the first few pronic numbers are:
𝑃 = 𝑛 ∗ (𝑛 + 1) 𝑛
𝑃1 = 1 ∗ (1 + 1) = 2
𝑃2 = 2 ∗ (2 + 1) = 6
𝑃3 = 3 ∗ (3 + 1) = 12
𝑃4 = 4 ∗ (4 + 1) = 20

````python
for i in range(1,101):
    if i*(i+1)<=100:
        print(i*(i+1))
````
